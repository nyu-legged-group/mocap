```python
import os
currentdir=os.getcwd()
parentdir = os.path.dirname(currentdir)
os.sys.path.insert(1, parentdir+'/src')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
com_filename = parentdir+'/data/ddp/com_data.csv'
com = np.loadtxt(com_filename, delimiter=',')


rg_vel_filename = parentdir+'/data/ddp/bullet_gripper_vel_time_norm.csv'
rg_vel = np.loadtxt(rg_vel_filename, delimiter=',', skiprows=1)

gnf_l_filename = parentdir+'/data/ddp/bullet_l_gnf.csv'
gnf_l = np.loadtxt(gnf_l_filename, delimiter=',')

gnf_r_filename = parentdir+'/data/ddp/bullet_r_gnf.csv'
gnf_r = np.loadtxt(gnf_r_filename, delimiter=',')

am_filename = parentdir+'/data/ddp/am_data.csv'
am = np.loadtxt(am_filename, delimiter=',')
```

```python
plt.figure(dpi=300)
plt.plot(com[:1000,0], com[:1000,3], label='CoM Height',lw=5, c='g',alpha =.5)
plt.ylabel('CoM height [m]')
plt.xlabel('time [s]')
plt.show()
```

```python
plt.figure(dpi=300)
plt.plot(rg_vel[:5000,0], rg_vel[:5000,1],lw=3, c='r',alpha =.5)
plt.ylabel('Right gripper velocity [m/s]')
plt.xlabel('time [s]')
plt.show()
```

```python
plt.figure(dpi=300)
plt.plot(np.arange(len(gnf_r))*0.001, gnf_r[:5000], label= "right foot", lw=2, c='g',alpha =.5)
plt.ylabel('Ground reaction force [m]')
plt.xlabel('time [s]')
plt.show()
```

```python
plt.figure(dpi=300)

am_norm = []
for i in range(am.shape[0]):
    norm = np.sqrt(am[i,1]**2 + am[i,2]**2+am[i,3]**2)
    am_norm.append(norm)
    
# plt.plot(am[:1000,0], am[:1000, 1], label= "x", lw=2, c='g',alpha =.5)
# plt.plot(am[:1000,0], am[:1000, 2], label= "y", lw=2, c='r',alpha =.5)
# plt.plot(am[:1000,0], am[:1000, 3], label= "z", lw=2, c='k',alpha =.5)
plt.plot(am[:1000,0], am_norm[:1000], label= "b", lw=3, c='b',alpha =.5)
# plt.legend()
plt.ylabel(r'Angular momentum [${\rm kg m^2/s}$]')
plt.xlabel('time [s]')
plt.show()
```
